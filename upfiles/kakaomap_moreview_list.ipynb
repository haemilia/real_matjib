## 카카오맵 일반음식점 상세보기 주소 크롤링
* 마포구 일반음식점 중 연남동 소재 일반음식점들만 추려냄
* 연남동 음식점들의 카카오맵 '상세보기' 주소 가져와 정리

```python
import pandas as pd
import numpy as np

# CSV 파일 불러오기
df = pd.read_csv('seoul_mapogu_general_restaurants.csv', encoding='cp949')
df
```

```python
## 전체 컬럼명 확인
df.columns
```

```python
## 필요한 컬럼만 선택해 작업용 DF 만들기
filtered_df = df[['사업장명', '소재지전체주소', '도로명전체주소', '좌표정보X(EPSG5174)', '좌표정보Y(EPSG5174)']]
filtered_df
```

```python
# '소재지전체주소' 기준으로 '연남동'이 포함된 행만 추출
filtered_df = filtered_df[filtered_df['소재지전체주소'].str.contains('연남동', na=False)]
filtered_df
```

```python
## 세 개의 컬럼 데이터(사업장명, x좌표, y좌표)가 동일한 경우 첫 번째 값만 유지하고 나머지는 중복으로 제거 
filtered_df = filtered_df.drop_duplicates(subset=['사업장명', '좌표정보X(EPSG5174)', '좌표정보Y(EPSG5174)'], keep='first')
filtered_df
```

```python
## '소재지전체주소' 컬럼 정리 (번지수 까지만 남기기)
filtered_df['소재지전체주소'] = filtered_df['소재지전체주소'].apply(lambda x: ' '.join(x.split()[:4]))

## '상세보기주소' 컬럼 추가 (NaN 값으로 채움)
filtered_df['상세보기url'] = np.nan

# 결과 출력
filtered_df
```

```python
## 상세보기주소 크롤링
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import csv
import pandas as pd
import numpy as np
import re

import warnings
warnings.filterwarnings('ignore')

url = 'https://map.kakao.com/'
driver = webdriver.Chrome()
driver.get(url)

result_url_list = []

from googletrans import Translator

## 검색에 필요한 상호명 처리 함수
def process_string(text):    
    if not text.strip():
        # print("입력된 문자열이 없습니다!")
        cleaned_text = "없음"
        return cleaned_text

    korean_pattern = re.compile(r'[\uAC00-\uD7A3]+')
    has_korean = korean_pattern.search(text)

    if not has_korean:
        translator = Translator()
        cleaned_text = translator.translate(text, dest='ko').text
        # print("한글이 없습니다!")
        return cleaned_text
    else:
        # 영문, 숫자, 한문은 삭제, & 및 괄호 제외 특수기호는 빈칸으로 대체        
        temp_text = re.sub(r'[()]', ' ', text)
        temp_text = re.sub(r'&', ' ', temp_text)
        temp_text = re.sub(r'[^가-힣\s]+', '', temp_text)
        cleaned_text = re.sub(r'\s+', ' ', temp_text).strip()
        # print("변환 결과:", cleaned_text)
        return cleaned_text

## 검색 결과('장소' 리스트)에서 상호에 해당하는 '상세보기' 링크 가져오는 함수
def find_moreview_url(soup):
    global serch_shop_name
    # 검색할 상호명서 영자, 한자, 숫자, 특수기호 제거하기
    # cleaned_text = re.sub(r'[a-zA-Z\u4E00-\u9FFF0-9\W_]+', ' ', serch_shop_name)
    # target = cleaned_text.split()[0]
    target = serch_shop_name.split()[0]
            
    link_name_links = soup.find_all('a', class_="link_name")

    for link_name_link in link_name_links:
        if link_name_link.string and len(set(target) & set(link_name_link.string)) >= 2:
            # 현재 'link_name' 태그의 다음 요소부터 탐색
            next_element = link_name_link.find_next()
            while next_element:
                if next_element.name == 'a' and 'moreview' in next_element.get('class', []):
                    if 'href' in next_element.attrs:
                        return next_element['href']
                    else:
                        # 'moreview' 클래스를 가진 <a> 태그지만 href가 없는 경우
                        return np.nan
                next_element = next_element.find_next()

    return np.nan # 조건을 만족하는 'moreview' 링크를 찾지 못한 경우

for i in range(len(filtered_df)):
    # serch_shop_name = filtered_df['사업장명'].iloc[i]
    serch_shop_name = process_string(filtered_df['사업장명'].iloc[i])
    serch_shop_addres = filtered_df['소재지전체주소'].iloc[i]       

    search_area = driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]')   # 검색창 선택
    search_area.clear()     # 검색창 비우기

    ## 검색어 넣고 엔터
    search_area.send_keys(serch_shop_addres)
    # search_area.send_keys(serch_shop_addres.split()[:4])  # DF서 주소명 정제 안했을 경우
    time.sleep(0.2)
    driver.find_element(By.XPATH, '//*[@id="search.keyword.submit"]').send_keys(Keys.ENTER)
    time.sleep(2)

    ## 검색 결과에서 '장소' 버튼 클릭
    driver.find_element(By.XPATH, '//*[@id="info.main.options"]/li[2]/a').send_keys(Keys.ENTER)
    time.sleep(0.2)

    ## 페이지 내용 파싱
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    moreview_url = find_moreview_url(soup)
    
    print('상호명=', serch_shop_name, ' /  상세보기URL=', moreview_url)
    result_url_list.append(moreview_url)
    # filtered_df['상세보기url'].iloc[i] = moreview_url    

print('**크롤링 완료**')
driver.quit()
```
